```python
# load csv file using np
import numpy as np
import pandas as pd

filename = "AAPL.csv"
filepath = "./dataset/" + filename
data = pd.read_csv(filepath)

print(data.head())
```

```python
def calculate_differences(csv_file):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(csv_file)

    # extract first column
    first_column = df.iloc[:, 0]

    # remove first column
    df = df.iloc[:, 1:]

    # Calculate the differences
    differences = df.diff()

    # Add the first column back to the DataFrame
    differences.insert(0, 'date', first_column)

    # remove first row
    differences = differences.iloc[1:, :]

    # Return the DataFrame with differences
    return differences
```

```python
def calculate_differences_in_percentage(csv_file):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(csv_file)

    # extract first column
    first_column = df.iloc[:, 0]

    # remove first column
    df = df.iloc[:, 1:]

    # Calculate the differences
    differences = df.pct_change()

    # Add the first column back to the DataFrame
    differences.insert(0, 'date', first_column)

    # remove first row
    differences = differences.iloc[1:, :]

    # Return the DataFrame with differences
    return differences
```

```python
data_diff = calculate_differences(filepath)
data_diff_pct = calculate_differences_in_percentage(filepath)
```

```python
# plot both in seperate graph

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(data['OT'], label='AAPL')
plt.title('AAPL Close Price')
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(data_diff['OT'], label='AAPL')
plt.title('AAPL Close Price Difference')
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(data_diff_pct['OT'], label='AAPL')
plt.title('AAPL Close Price Difference in Percentage')
plt.legend()
plt.show()
```
