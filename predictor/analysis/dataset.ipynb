```python
# Basic Data Analysis

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# get data
data_path = "../dataset/"
data = pd.read_csv(data_path + 'AAPL.csv')
data['Timestamp'] = pd.to_datetime(data['Timestamp'])

data.head()
```

```python

year_data = data[data['Timestamp'].dt.year == 2022]
month_data = data[data['Timestamp'].dt.month == 3]
day_data = data[data['Timestamp'] >= '2022-03-01']
day_data = day_data[day_data['Timestamp'] <= '2022-03-02']
```

```python
# plot OHLC



fig = plt.figure(figsize=(10, 5))

# Plot OHLC
plt.plot(day_data["Timestamp"], day_data['Open'], label='Open')
plt.plot(day_data["Timestamp"], day_data['High'], label='High')
plt.plot(day_data["Timestamp"], day_data['Low'], label='Low')
plt.plot(day_data["Timestamp"], day_data['Close'], label='Close')

# Plot Volume as bar graph with secondary y-axis
ax2 = plt.twinx()
ax2.set_ylim([0, 20000000])
ax2.bar(day_data["Timestamp"], day_data['Volume'], label='Volume', alpha=0.8, width=0.008)

plt.title('2022-03-01 AAPL OHLC')
plt.xlabel('Date')
plt.ylabel('Price')

# Set legends for both plots
lines, labels = fig.axes[0].get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
plt.legend(lines + lines2, labels + labels2)

plt.show()
```

```python
fig = plt.figure(figsize=(10, 5))

plot_data = day_data
title_date = '2022-03-01' if plot_data is day_data else '2022-03'

# Plot OHLC
# plt.plot(plot_data['Timestamp'], plot_data['SMA_100'], label='SMA_100')
# plt.plot(plot_data['Timestamp'], plot_data['SMA_50'], label='SMA_50')
plt.plot(plot_data['Timestamp'], plot_data['SMA_10'], label='SMA_10')
# plt.plot(plot_data['Timestamp'], plot_data['EMA_100'], label='EMA_100')
# plt.plot(plot_data['Timestamp'], plot_data['EMA_50'], label='EMA_50')
plt.plot(plot_data['Timestamp'], plot_data['EMA_10'], label='EMA_10')
plt.plot(plot_data['Timestamp'], plot_data['Close'], label='Close')

plt.title(title_date + ' AAPL Close & SMA & EMA')
plt.xlabel('Date')
plt.ylabel('Price')

# Set legends for both plots
lines, labels = fig.axes[0].get_legend_handles_labels()
plt.legend()

plt.show()
```

```python
fig = plt.figure(figsize=(10, 5))

plot_data = month_data
title_date = '2022-03-01' if plot_data is day_data else '2022-03'

# Plot OHLC
# plt.plot(plot_data['RSI_100'], label='RSI_100')
plt.plot(plot_data['RSI_50'], label='RSI_50')
# plt.plot(plot_data['RSI_10'], label='RSI_10')

ax2 = plt.twinx()
ax2.plot(plot_data['Close'], label='Close', color='orange')

plt.title(title_date + ' AAPL Close & RSI')
plt.xlabel('Date')
plt.ylabel('Price')

# Set legends for both plots
lines, labels = fig.axes[0].get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
plt.legend(lines + lines2, labels + labels2)

plt.show()
```
